#checking the badlines

```python
bad_lines = []

# Define a function to collect the problematic rows
def collect_bad_lines(bad_line_list):
    bad_lines.append(bad_line_list)
    return None  # Returning None tells Pandas to skip the line after capturing it
```

```python
# Load the dataset using the callback
df = pd.read_csv('/content/drive/MyDrive/8th Semester (final project )/NewFeatures/Sentiment Analysis/DataSets/train.csv',
                 on_bad_lines=collect_bad_lines,
                 engine='python')

# Print how many bad lines were found
print(f"Total problematic lines found: {len(bad_lines)}")
```

```python
# Load the dataset using the callback
df1 = pd.read_csv('/content/drive/MyDrive/8th Semester (final project )/NewFeatures/Sentiment Analysis/DataSets/test.csv',
                 on_bad_lines=collect_bad_lines,
                 engine='python')

# Print how many bad lines were found
print(f"Total problematic lines found: {len(bad_lines)}")
```

#finding Other language

```python
#!pip install langdetect
from langdetect import detect
def robust_lang_check(text):
    try:
        # Instead of just 200 chars, use up to 500 for better accuracy
        return detect(text[:500])
    except:
        return "unknown"
```

```python
df['detected_lang'] = df['text'].apply(robust_lang_check)
```

```python
# Keep only English rows
df = df[df['detected_lang'] == 'en'].copy()

# Double-check
print(df['detected_lang'].value_counts())
```

```python
df1['detected_lang'] = df1['text'].apply(robust_lang_check)
```

```python
# Keep only English rows
df1 = df1[df1['detected_lang'] == 'en'].copy()

# Double-check
print(df1['detected_lang'].value_counts())
```
